# Pandas 
Python for Data Analysis (Pandas) is a powerful library that allows using the Data Frame structure. This library simplifies statistical analysis by providing the DataFrame as a coherent strucutre with some preliminary statistical fitting and visualization tools. However for very large data sets, we recommend going back Numpy. Most of Pandas functionalities can be re-produced more efficiently with scipy, skitlearn, and numpy libraries. Pandas is only more convenient and saves more coding time.

# Random Number Generation
np.random includes some functions for random number generation. Below we try to generate 

```python
import numpy as np
import matplotlib.pyplot as plt

#this command is required for jupyter notebook, otherwise everytime you must use plt.show()
%matplotlib inline


n = 200

x_tr = np.linspace(0.0, 2.0, n)
y_tr = np.exp(3*x_tr)


mu, sigma = 0,50
np.random.seed(1)
y = y_tr + np.random.normal(loc=mu, scale= sigma, size=len(x_tr))


plt.plot(x_tr,y,".",mew=3);
plt.plot(x_tr, y_tr,"--r",lw=3);


plt.xlabel('Explanatory variable (x)')
plt.ylabel('Dependent variable (y)')
```

The red curve is defined by the function 

\begin{equation}
 \\  f(x) = e^{\ 3 x} \ \ ,
\end{equation}

and the blue dots are actually the dependent variable defined by 

\begin{equation}
    y = e^{\ 3 x} + \varepsilon \ \ \text{where}  \ \  \varepsilon \sim \mathcal{N}(0,50^2) \ \ .
\end{equation}
